Importing the Libraries

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('/content/min.jpg')
```

```python
# 1. Gaussian Filter
gaussian_blurred = cv2.GaussianBlur(img, (5, 5), 0)
```

```python
# 2. Median Filter
median_blurred = cv2.medianBlur(img, 5)
```

```python
# 3. Sobel Filter (Edge Detection)
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=5)
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=5)
sobel_combined = np.hypot(sobelx, sobely)
sobel_combined = np.uint8(sobel_combined)
```

```python
# 4. Box Filter (Averaging Filter)
box_blurred = cv2.blur(img, (5, 5))
```

```python
# 5. Laplacian Filter (Edge Detection)
laplacian = cv2.Laplacian(img, cv2.CV_64F)
```

```python
# Display the results (optional, but recommended)
plt.figure(figsize=(15,10))
```

```python
plt.imshow(img)
plt.title('Original')
plt.axis('off')
```

```python
plt.imshow(gaussian_blurred)
plt.title('Gaussian')
plt.axis('off')
```

```python
plt.imshow(median_blurred)
plt.title('Median')
plt.axis('off')
```

```python
plt.imshow(sobel_combined)
plt.title('Sobel')
plt.axis('off')
```

```python
plt.imshow(box_blurred)
plt.title('Box')
plt.axis('off')
```

```python
plt.imshow(laplacian)
plt.title('Laplacian')
plt.axis('off')
```

```python
def add_gaussian_noise(image, mean=0, std_dev=25):
  noise = np.random.normal(mean, std_dev, image.shape).astype(np.uint8)
  noisy_image = cv2.add(image, noise)

  return noisy_image
```

```python
add_gaussian_noise(img)
```
